# Dicionário de Dados: Consumo Mundial de Energia

## Descrição
Este conjunto de dados apresenta o consumo mundial de energia por diferentes fontes, com dados anuais de 1983 a 2022.

## Variáveis

| Coluna | Tipo | Descrição |
|--------|------|-----------|
| Year | Integer | Ano de referência dos dados |
| Other_renewables | Float | Consumo de energia de outras fontes renováveis (TWh) |
| Biofuels | Float | Consumo de energia proveniente de biocombustíveis (TWh) |
| Solar | Float | Consumo de energia solar (TWh) |
| Wind | Float | Consumo de energia eólica (TWh) |
| Hydropower | Float | Consumo de energia hidrelétrica (TWh) |
| Nuclear | Float | Consumo de energia nuclear (TWh) |
| Gas | Float | Consumo de energia proveniente de gás natural (TWh) |
| Oil | Float | Consumo de energia proveniente de petróleo (TWh) |
| Coal | Float | Consumo de energia proveniente de carvão (TWh) |
| Traditional_biomass | Float | Consumo de energia proveniente de biomassa tradicional (TWh) |

## Observações:
- Unidade de medida: Terawatt-hora (TWh)
- Período: 40 anos (1983-2022)
- Frequência: Anual
- Os dados apresentam valores decimais com até 6 casas de precisão
- O conjunto de dados não apresenta valores nulos

## Fonte dos Dados
Este arquivo está localizado em: 1-substituicaoConsumoDeEnergiaMundial.csv

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("./data/substituicaoConsumoDeEnergiaMundial.csv")
df
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df["Year"], df["Oil"], label="Oil energy")
plt.xlabel("Year")
plt.ylabel("Consumption (TWh)")
plt.title("Oil Consumption")
plt.legend()
plt.show()
```

## Análise Rápida: Consumo de Petróleo (1983-2022)

O gráfico mostra o consumo global de petróleo em Terawatts-hora (TWh) de 1983 a 2022. Observa-se um aumento geral com algumas flutuações, incluindo uma queda notável em 2020, provavelmente devido à pandemia de COVID-19, seguida por uma recuperação parcial. O crescimento é mais acentuado a partir de meados dos anos 2000.

```python
plt.figure(figsize=(12, 6))
plt.plot(df["Year"], df["Gas"], label="Gas energy")
plt.xlabel("Year")
plt.ylabel("Consumption (TWh)")
plt.title("Gas Consumption")
plt.legend()
plt.show()
```

## Análise Rápida: Consumo de Energia Eólica (1983-2022)

O gráfico mostra o consumo global de energia eólica em Terawatts-hora (TWh) de 1983 a 2022. O consumo é praticamente nulo até o final dos anos 1990, com um aumento gradual no início dos anos 2000 e um crescimento exponencial a partir de 2010. Este aumento significativo indica um rápido desenvolvimento e adoção da energia eólica como fonte de energia nas últimas décadas.

```python
plt.figure(figsize=(12, 6))
plt.plot(df["Year"], df["Biofuels"], label="Biofuels energy")
plt.plot(df["Year"], df["Solar"], label="Solar energy")
plt.plot(df["Year"], df["Wind"], label="Wind energy")
plt.plot(df["Year"], df["Hydropower"], label="Hidropower energy")
plt.plot(df["Year"], df["Nuclear"], label="Nuclear energy")
plt.plot(df["Year"], df["Gas"], label="Gas energy")
plt.plot(df["Year"], df["Oil"], label="Oil energy")
plt.plot(df["Year"], df["Coal"], label="Coal energy")
plt.plot(df["Year"], df["Traditional_biomass"], label="Traditional biomass energy")
plt.xlabel("Year")
plt.ylabel("Consumption (TWh)")
plt.title("World energy consumption by source")
plt.legend()
plt.show()
```

## Análise Rápida: Consumo Global de Energia por Fonte (1983-2022)

O gráfico compara o consumo global de diversas fontes de energia em Terawatts-hora (TWh) ao longo de quase quatro décadas. Observa-se:

* **Domínio dos Combustíveis Fósseis**: Petróleo e gás natural mantêm-se como as maiores fontes de energia, com carvão em um distante terceiro lugar.
* **Crescimento das Renováveis**: Solar e eólica apresentam um crescimento exponencial a partir de meados dos anos 2000, embora ainda representem uma fração menor do consumo total.
* **Biomassa Tradicional Decrescente**: O consumo de biomassa tradicional diminui ao longo do tempo, indicando mudanças nos padrões de uso de energia.
* **Estabilidade Nuclear e Hidrelétrica**: A energia nuclear e hidrelétrica mostram um crescimento constante, embora moderado, ao longo do período.

Em resumo, o gráfico destaca a transição energética em curso, com um aumento notável nas energias renováveis, mas ainda sob o domínio dos combustíveis fósseis.

```python
df['Fossil Fuels'] = df['Oil'] + df['Gas'] + df['Coal']
df['Renewables'] = df['Biofuels'] + df['Solar'] + df['Wind'] + df['Hydropower']

plt.figure(figsize=(12, 6))
plt.plot(df['Year'], df['Fossil Fuels'], label='Fossil Fuels')
plt.plot(df['Year'], df['Renewables'], label='Renewables')
plt.plot(df['Year'], df['Nuclear'], label='Nuclear')
plt.xlabel("Year")
plt.ylabel("Consumption (TWh)")
plt.title("Energy Consumption by Category")
plt.legend()
plt.show()
```
